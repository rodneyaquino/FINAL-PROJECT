### ELBOW METHOD 

```python
# Elbow Method is a for loop that trains various KNN models with different k values, 
# then keeps track of the error_rate for each of these models with a list.
# REFERENCE | GitHub | analyticalmonk | https://github.com/analyticalmonk/KMeans_elbow
from sklearn.cluster import KMeans

# Try knn values 1 through 40 and check error rate for each
# goal is to find best knn value

def elbow_plot(data, maxK=10, seed_centroids=None):
    """
        parameters:
        - data: pandas DataFrame (data to be fitted)
        - maxK (default = 10): integer (maximum number of clusters with which to run k-means)
        - seed_centroids (default = None ): float (initial value of centroids for k-means)
    """
    sse = {} # create empty list to store results
    print("PROCESSING K VALUES 1 to 20, PLEASE STAND BY")
    
    for k in range(1, 20): 
        # Try knn values 1 through 20 and check error rate for each
        # goal is to find best knn value
              
        if seed_centroids is not None:
            seeds = seed_centroids.head(k)
            kmeans = KMeans(n_clusters=k, max_iter=500, n_init=100, 
                            random_state=0, init=np.reshape(seeds, (k,1))).fit(data)

        else:
            kmeans = KMeans(n_clusters=k, max_iter=300, n_init=100, random_state=0).fit(data)

        # Inertia: Sum of distances of samples to their closest cluster center
        sse[k] = kmeans.inertia_

    #PLOT the GRAPH
    plt.figure(figsize=(20,6)) # set plot figure size
    plt.plot(list(sse.keys()), list(sse.values()),color='blue',linestyle='dashed',marker='o',
         markerfacecolor='red',markersize=10) # set graph parameters
    plt.title('Error Rate vs K value') # set Title of graph
    plt.xlabel('K') # set label for X
    plt.ylabel('Error Rate') # set label for Y
    return

elbow_plot(X) # initiate elbow
```

```python
# KNN K-NEAREST NEIGHGORS PART II | USING NEW K VALUE 
# Looks like the best K value is at the Elbow Bend. Which is around 3. 
# With that Knowledge, we will set the K value to 3 and see if it improves our Accuracy Score

model_knn = KNeighborsClassifier(n_neighbors = 3) #create KNeighborsClassifier object variable SETTING K to 3
model_knn.fit(X_train, y_train) #train data for KNeighborsClassifier model
```

```python
# KNN K-NEAREST NEIGHGORS PART II | MODEL SCORE
model_knn.score(X_test,y_test) #test accuracy
```
